```python
#Deciding K value
X1=df[['Age', 'Spending Score (1-100)']].iloc[:,:].values
inertia = []
for n in range(1, 15):
  algorithm = (KMeans(n_clusters = n, init='k-means++', n_init =10, max_iter = 300, tol=0.0001,
                      random_state=111, algorithm ='elkan'))
  algorithm.fit(X1)
  inertia.append(algorithm.inertia_)
```

```python
plt.figure(1, figsize=(15,6))
plt.plot(np.arange(1,15), inertia, 'o')
plt.plot(np.arange(1,15), inertia, '-', alpha=0.5)
plt.xlabel('Number of Clusters'), plt.ylabel('Inertia')
plt.show()
```

```python
#Applying KMeans for k=4
algorithm = (KMeans(n_clusters = n, init='k-means++', n_init =10, max_iter = 300, tol=0.0001,
                      random_state=111, algorithm ='elkan'))
algorithm.fit(X1)
labels1=algorithm.labels_
centroids1 = algorithm.cluster_centers_
```

```python
h=0.02
x_min, x_max = X1[:,0].min() - 1, X1[:,0].max() + 1
y_min, y_max = X1[:,0].min() - 1, X1[:,0].max() + 1
xx, yy =np.meshgrid(np.arange(x_min, x_max,h), np.arange(y_min, y_max, h))
Z = algorithm.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
plt.figure(1, figsize=(15,7))
plt.clf()
Z=Z.reshape(xx.shape)
plt.imshow(Z, interpolation ='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Pastel2, aspect = 'auto', origin = 'lower')

plt.scatter(x='Age', y='Spending Score (1-100)', data=df, c=labels1, s=100)
plt.scatter(x=centroids1[:,0], y=centroids[:,1], s=300, c='red', alpha=0.5)
plt.ylabel('Spending Score (1-100)'),plt.xlabel('Age')
plt.show()
```
